```python
import pandas as pd
from geopy.distance import lonlat, distance, geodesic
pd.options.display.max_columns = 100
```

```python
df = pd.read_csv('citypairs.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df['combined_pop'] = df['pop']+df['pop_x']
```

```python
df.sort_values(by='pop', ascending = False).drop_duplicates(subset='combined_pop').sort_values(by='combined_pop', ascending = False)
```

```python
geodesic(lonlat(df['lat'][24], df['lng'][24]), lonlat(df['lat_x'][0], df['lng_x'][0]))
```

```python
df['distance'] = [geodesic(lonlat(df['lng'][x], df['lat'][x]), lonlat(df['lng_x'][x], df['lat_x'][x])).km 
                  for x in range(df.shape[0])]
```

```python
df = df[df['distance'] > 0]
df.head()
```

```python
df.sort_values('distance', ascending = False)
```

```python
df['power'] = df['combined_pop'] / df['distance']
```

```python
df.sort_values(by='power', ascending = False).drop_duplicates(subset='power')
```

```python
df.to_csv('distances.csv')
```
